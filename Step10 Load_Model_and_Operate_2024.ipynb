## Define DNN Model

```python
def DNN_regression_model(L2_factor, drop_factor, train_shape_1, Leaky_ReLu):
    if Leaky_ReLu:
        activation = LeakyReLU()
    else:
        activation = 'relu'
    def Sub_Layers(N, x, dropout=True):
        x = Dense(N, activation=activation,
                  kernel_regularizer = regularizers.l2(L2_factor))(x)
        x = BatchNormalization()(x)
        if dropout:
            x = Dropout(drop_factor)(x)
        return(x)

    inputs = Input(shape = [train_shape_1,])
    x = Sub_Layers(128, inputs)
    x = Sub_Layers(64, x)
    x = Sub_Layers(64, x)
    x = Sub_Layers(32, x)
    x = Sub_Layers(16, x)
    x = Sub_Layers(8, x, dropout=False)
    x = Dense(8, activation=activation, kernel_regularizer = regularizers.l2(L2_factor))(x)
    
    outputs = Dense(1, activation = 'linear')(x)
    model = Model(inputs, outputs)
    model.compile(loss = 'mse', optimizer = 'rmsprop', metrics = ['mae'])
    return(model)
    
```

### Prediction Functions

```python
def predict(X, model, LRM, scaler):
    keras = 'keras' in str(type(model))
    if X.shape == 1:
        X = X.reshape(1, -1)
    if keras:
        y = model.predict(scaler(X),verbose=0)
    else:
        y = model.predict(scaler(X))
        y = y.reshape((-1, 1))
    y = LRM(y)
    return y

def ensemble_predict(X, models, weights, LRMs, scaler):
    ensemble_y = np.zeros((X.shape[0],))
    for model_name in models.keys():
        this_weight = weights[model_name]
        this_y = predict(X, models[model_name], LRMs[model_name].predict, scaler)
        ensemble_y += this_y * this_weight
    ensemble_y /= sum(list(weights.values()))
    return ensemble_y
```

### Functions

#### Name Converter

```python
def player_name_convert(name):
    if len(name.split(',')[1].split(' ')) > 2:
        return name.split(',')[0] + ', ' + name.split(',')[1].split(' ')[-1]
    return name
```

#### Get Data and Xy with Player ID

```python
def find_player_id(player_name, birth_year=None):
    player_name = player_name_convert(player_name)
    df = player_id_df[player_id_df['player_name'] == player_name]
    if not birth_year is None:
        df = df[df['birth_year']==birth_year]
    return df['player_id'].iloc[0]

def read_raw_data_for_given_id(player_id):    
    raw_data = {'player_id':player_id}
    for yr in raw_dfs.keys():
        this_series = raw_dfs[yr][raw_dfs[yr]['player_id']==player_id]
        if len(this_series) >= 1:
            raw_data[yr] = this_series        
    return raw_data

def get_XY_for_given_id_season(player_id, season):
    raw_data = read_raw_data_for_given_id(player_id)
    
    if not (season - 1 in raw_data.keys() and season - 2 in raw_data.keys()):
        print('There is no enough data of this player to do prediction.')
        return None
    
    enough_pa = True
    for i in range(1,3):
        try:
            pa = raw_data[season - i]['pa'].iloc[0]
        except:
            pa = 0
        if pa < 50:
            print(f'The PA of {player_names[0]} in {season - i} is only {pa} (<50).')
            enough_pa = False
    
    if not enough_pa:
        print(f'The models were trained base on data of players who got PA >= 50\
in the two seasons continuously before the season we want to predict. {player_names[0]} \
did not got enough PAs, so the error of the prediction might be larger than expectation.')
    
    col_list_A = column_df[column_df['Season_offset'] == -2]['Column'].tolist()
    col_list_B = column_df[column_df['Season_offset'] == -1]['Column'].tolist()
    col_list_C = column_df[column_df['Season_offset'] == 0]['Column'].tolist()
    
    X = np.array(raw_data[season - 2][col_list_A])
    X = np.concatenate((X, np.array(raw_data[season - 1][col_list_B])), axis = 1)
    
    X_c = [[],]
    for col in col_list_C:
        if col[0:2] == 'is':
            this_y = int(col[2::])
            if season == this_y:
                X_c[0].append(1)
            else:
                X_c[0].append(0)
        elif col[0:2] == 'DH':
            last_DH = raw_data[season-1]['isDH'].iloc[0]
            if season >= 2022:
                this_DH = 1
            elif not season in raw_data.keys():
                this_DH = last_DH
            else:
                this_DH = raw_data[season]['isDH'].iloc[0]
            X_c[0].append(this_DH - last_DH)
    
    X = np.concatenate((X, X_c), axis=1)
    X[np.isnan(X)] = 0
    
    #-------------If the season we want to predict has not finished yet---------
    if not season in raw_dfs.keys():
        return X, np.array([None])
    
    y_previous = raw_data[season-1]['wRAA'].iloc[0]
    if season in raw_data.keys():
        y_now = raw_data[season]['wRAA'].iloc[0]
    else:
        y_now = 0
    
    y = y_now - y_previous
    y = np.array([y])
    return X, y

def read_data_table_for_single_ID(player_id, years=years,
                                  columns=['Age', 'PA', 'AB', 'R', 'H', '2B', '3B',
                                           'HR', 'RBI', 'BB', 'SO', 'HBP',
                                           'AVG', 'OBP', 'SLG', 'OPS'],
                                  wRAA=True):
    
    result_dict = {}
    result_dict['Year'] = []
    for col in columns:
        result_dict[col] = []
    
    if wRAA:
        result_dict['wRAA'] = []
    for yr in years:
        this_df = first_data_dfs[yr]
        this_df = this_df[this_df['player_id']==player_id]
        if len(this_df) > 0:
            result_dict['Year'].append(yr)
            if wRAA:
                try:
                    this_wRAA = raw_dfs[yr][raw_dfs[yr]['player_id']==player_id]['wRAA'].iloc[0]
                except:
                    this_wRAA = None
                result_dict['wRAA'].append(this_wRAA)          
            for col in columns:
                result_dict[col].append(this_df[column_mapping[col]].iloc[0])
    return pd.DataFrame(result_dict)
```

#### Get Data and Xy with Player Name

```python
def read_data_table_for_single_player(player_name,
                                      birth_year=None,
                                      columns=['Age', 'PA', 'AB', 'R', 'H', '2B', '3B',
                                               'HR', 'RBI', 'BB', 'SO', 'HBP',
                                               'AVG', 'OBP', 'SLG', 'OPS'],
                                      wRAA=True):
    player_name = player_name_convert(player_name)
    player_id = find_player_id(player_name, birth_year=birth_year)
    return read_data_table_for_single_ID(player_id, columns=columns, wRAA=wRAA)

def get_XY_for_given_player_season(player_name, season, birth_year=None):
    player_name = player_name_convert(player_name)
    player_id = find_player_id(player_name, birth_year=birth_year)
    return get_XY_for_given_id_season(player_id, season)

def get_XY_for_one_player(player_name, birth_year=None):
    player_name = player_name_convert(player_name)
    pa_table = read_data_table_for_single_player(player_name, columns=['PA'], wRAA=False)
    enough_pa_years = [pa_table['Year'].tolist()[i] for i in range(len(pa_table))\
                       if pa_table['PA'].tolist()[i] >= 50]
    
    if len(enough_pa_years) < 2:
        print('No enough PAs.')
        return
    
    feasible_years = []
    count = 0
    for yr in range(enough_pa_years[0], enough_pa_years[-1] + 1):
        if yr in enough_pa_years:
            count += 1
        else:
            count = 0
        if count >= 2:
            feasible_years.append(yr+1)
    #print(feasible_years)
    if len(feasible_years) == 0:
        print('No continuously enough PAs.')
        return
    
    X_list = []
    y_list = []
    for yr in feasible_years:
        this_X, this_y = get_XY_for_given_player_season(player_name, yr)
        X_list.append(this_X)
        y_list.append(this_y)
    
    X = np.concatenate(X_list, axis=0)
    y = np.concatenate(y_list, axis=0)
    return X, y, feasible_years
```

#### Get Xy by season(s)

```python
def get_player_list(season, player_set):
    # player_set could be 'train', 'test', or 'all'
    if season == 2024:
        player_idx = list(player_df_new_year['player_id'])
        player_list = list(player_df_new_year['player_name'])
        player_df_this_year = player_df_new_year
    else:
        if player_set == 'train':
            this_player_season = train_player_season_df[train_player_season_df['season'] == season]
        if player_set == 'test':
            this_player_season = test_player_season_df[test_player_season_df['season'] == season]
        if player_set == 'all':
            this_player_season = pd.concat(
                (train_player_season_df, test_player_season_df))
            this_player_season = this_player_season[this_player_season['season'] == season]
        player_idx = list(set(this_player_season['player_id']))

        player_list_this_year = {'player_id': [], 'player_name': []}
        for ID in player_idx:
            player_list_this_year['player_id'].append(ID)
            player_name = player_id_df[player_id_df['player_id']
                                       == ID]['player_name'].iloc[-1]
            player_list_this_year['player_name'].append(player_name)
        player_df_this_year = pd.DataFrame(
            player_list_this_year).sort_values('player_name').reset_index()
        player_list = list(player_df_this_year['player_name'])
    return player_df_this_year, player_list

def read_data_table_for_one_season(season,
                                   player_set='all', # player_set could be 'train', 'test', or 'all'
                                   columns=['Age', 'PA', 'AB', 'R', 'H', '2B', '3B',
                                            'HR', 'RBI', 'BB', 'SO', 'HBP',
                                            'AVG', 'OBP', 'SLG', 'OPS'],
                                   wRAA=True,
                                   sorted_by='player_name',
                                   ascending=False,
                                   next_year_players=False):
    if next_year_players:
        player_df_this_year, player_list = get_player_list(season+1, player_set)
    else:
        player_df_this_year, player_list = get_player_list(season, player_set)
        
    dfs = []
    for i, ID in enumerate(player_df_this_year['player_id']):
        this_data_df = read_data_table_for_single_ID(ID, years=[season], columns=columns)
        this_data_df.insert(loc=0, column='Name', value=player_list[i])
        dfs.append(this_data_df)
    result_df = pd.concat(dfs, axis=0, ignore_index=True)
    if not sorted_by == 'player_name':
        result_df = result_df.sort_values(sorted_by, ascending=ascending).reset_index(drop=True)
    return result_df
    


def get_XY_for_single_season(season, player_set):
    # player_set could be 'train', 'test', or 'all'
    player_df_this_year, player_list = get_player_list(season, player_set)

    for i, ID in enumerate(player_df_this_year['player_id']):
        this_X, this_y = get_XY_for_given_id_season(ID, season)
        if i == 0:
            X = this_X
            y = this_y
            continue
        X = np.concatenate((X, this_X), axis=0)
        y = np.concatenate((y, this_y), axis=0)
    return X, y, player_list


def get_XY_for_many_seasons(seasons, player_set):
    Xs = []
    ys = []
    player_season_dict = {'player': [], 'season': []}
    for i, season in enumerate(seasons):
        this_X, this_y, this_player_list = get_XY_for_single_season(
            season, player_set)
        Xs.append(this_X)
        ys.append(this_y)
        player_season_dict['player'] += this_player_list
        player_season_dict['season'] += [season] * this_X.shape[0]
    X = np.concatenate(Xs, axis=0)
    y = np.concatenate(ys, axis=0)
    player_season_df = pd.DataFrame(player_season_dict)
    return X, y, player_season_df
```

### Estimate the impact of each column

```python
from scipy.special import erf

X, y, player_season_df = get_XY_for_many_seasons([*range(2019, 2025)], 'all')
N = X.shape[0]
partial_differenial = {'col_name':[], 'mean_slope':[], 'std_slope':[], 'p-value':[],
                       'mean_slope_per_z':[], 'std_slope_per_z':[]}
for shifted_col in range(X.shape[1]):
    delta = 0.001
    
    col_name = column_df['Column'].iloc[shifted_col]
    if column_df.iloc[shifted_col]['Season_offset'] == -2:
        col_name += '_A'
    if column_df.iloc[shifted_col]['Season_offset'] == -1:
        col_name += '_B'
    
    shifted_step = np.std(X[:,shifted_col]) * delta
    shifted_X_A = X * 1.0
    shifted_X_B = X * 1.0
    shifted_X_A[:, shifted_col] += shifted_step
    shifted_X_B[:, shifted_col] -= shifted_step

    #y = ensemble_predict(X, models, weights, LRMs, scaler.transform)
    shifted_y_A = ensemble_predict(shifted_X_A, models, weights, LRMs, scaler.transform)
    shifted_y_B = ensemble_predict(shifted_X_B, models, weights, LRMs, scaler.transform)

    y_diff = (shifted_y_A - shifted_y_B)/2
    mean_delta_y = np.mean(y_diff)
    std_delta_y = np.std(y_diff)

    p = 1 - erf(np.sqrt(0.5)*abs(mean_delta_y/std_delta_y)*np.sqrt(N-1))
    
    partial_differenial['col_name'].append(col_name)
    partial_differenial['mean_slope_per_z'].append(mean_delta_y/delta)
    partial_differenial['std_slope_per_z'].append(std_delta_y/delta)
    partial_differenial['mean_slope'].append(mean_delta_y/shifted_step)
    partial_differenial['std_slope'].append(std_delta_y/shifted_step)
    partial_differenial['p-value'].append(p)
    
#     if p < 0.05:
#         print('='*30)
#         print(column_df.iloc[shifted_col])
#         print(f'{mean_delta_y * 1000:.3f}')
#         print(f'{std_delta_y/np.sqrt(N-1) * 1000:.3f}')
#         print(f'{p:.3f}')
```

```python
partial_differenial_df = pd.DataFrame(partial_differenial)
partial_differenial_df.insert(loc=0, column='Temp', value=np.abs(partial_differenial_df['mean_slope_per_z']))
partial_differenial_df = partial_differenial_df.sort_values('Temp', ascending=False).reset_index(drop=True)
partial_differenial_df = partial_differenial_df.drop('Temp', axis=1)
```
