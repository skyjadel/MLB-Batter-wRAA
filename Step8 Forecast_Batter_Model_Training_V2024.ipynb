## Statistic to measure the performance of baseball players

### Weighted on-base average (wOBA)  </br>
<I>In baseball, wOBA (/'woʊbə/, or weighted on-base average) is a statistic, based on linear weights, designed to measure a player's overall offensive contributions per plate appearance. It is formed from taking the observed run values of various offensive events, dividing by a player's plate appearances, and scaling the result to be on the same scale as on-base percentage. Unlike statistics like OPS, wOBA attempts to assign the proper value for each type of hitting event. It was created by Tom Tango and his coauthors for The Book: Playing the Percentages in Baseball. (From Wikipedia)</I>

### Weighted on-base above replacement (wOBR)  </br>
Here we take wOBR = (wOBA - replacement wOBA) * PA to evaluate the overall contribution of a player. A high wOBR for a batter means that he or she is more likely to help their team to earn more runs. On the other hand, a lower wOBR for a pitcher means that he or she is more likely to save more runs for their team.</br></br>
The 'replacement wOBA' means the wOBA of a replacable player. Operationally, we choose the wOBA of the 75%th player of entire league as the 'replacement wOBA'.

```python
wobas = df['woba']; pas = df['pa']
sorted_wobas = np.sort(np.array(df_enough_pa['woba']))
replacement_woba = sorted_wobas[round(df_enough_pa.shape[0]*0.25)]

fig, ax = plt.subplots()
ax.hist((wobas - replacement_woba)*pas, 50)
ax.set_xlabel('wOBR')
ax.set_ylabel('counts')
```

```python
def plot_hist_with_many_text(y, title='', xlabel='', ylabel='', fig_size = (5,4), text_start_pos=(0.55, 0.8)):
    fig, ax = plt.subplots()
    fig.set_size_inches(fig_size)
    h = ax.hist(y, 50)
    h = ax.set_xlabel(xlabel)
    h = ax.set_ylabel(ylabel)
    h = ax.set_title(title)

    mean_data = np.mean(y)
    median_data = np.median(y)
    std_data = np.std(y)
    Q1_data = np.sort(y)[int(len(y)*0.25)]
    Q3_data = np.sort(y)[int(len(y)*0.75)]
    top10_data = np.sort(y)[int(len(y)*0.90)]
    top5_data = np.sort(y)[int(len(y)*0.95)]
    bottom10_data = np.sort(y)[int(len(y)*0.10)]

    x_lim = ax.get_xlim()
    y_lim = ax.get_ylim()

    def get_text_XY(Xr, Yr, x_lim=x_lim, y_lim=y_lim):
        x = (x_lim[1] - x_lim[0]) * Xr + x_lim[0]
        y = (y_lim[1] - y_lim[0]) * Yr + y_lim[0]
        return (x,y)

    line_interval = 0.06
    first_line_y = text_start_pos[1]
    line_start_x = text_start_pos[0]
    ax.text(*get_text_XY(line_start_x, first_line_y), f'Mean: {mean_data:.2f}')
    ax.text(*get_text_XY(line_start_x, first_line_y - 1 * line_interval), f'Std: {std_data:.2f}')
    ax.text(*get_text_XY(line_start_x, first_line_y - 2 * line_interval), f'10th percentile: {bottom10_data:.2f}')
    ax.text(*get_text_XY(line_start_x, first_line_y - 3 * line_interval), f'25th percentile: {Q1_data:.2f}')
    ax.text(*get_text_XY(line_start_x, first_line_y - 4 * line_interval), f'50th percentile: {median_data:.2f}')
    ax.text(*get_text_XY(line_start_x, first_line_y - 5 * line_interval), f'75th percentile: {Q3_data:.2f}')
    ax.text(*get_text_XY(line_start_x, first_line_y - 6 * line_interval), f'90th percentile: {top10_data:.2f}')
    ax.text(*get_text_XY(line_start_x, first_line_y - 7 * line_interval), f'95th percentile: {top5_data:.2f}')
```

## Select players with continous enough-PA seasons
For players with few PAs, many of the features we use here will be noisy due to lack of sample. So we only select players with enough PAs (100ish) per year to train the model.

```python
#----Find suitable players for analysis for each year
player_dict = {}
for year in years:    
    player_set_m1 = set(df_enough_pa[df_enough_pa.season_id.eq(year-1)]['player_id'])
    player_set_m2 = set(df_enough_pa[df_enough_pa.season_id.eq(year-2)]['player_id'])
    player_set_0 = set(df_enough_pa[df_enough_pa.season_id.eq(year)]['player_id'])
    if y_type == 'rate':
        #----Players must have 3 continous enough-PA seasons
        player_list = list(player_set_0.intersection(player_set_m1).intersection(player_set_m2))
    elif y_type == 'counts':
        if single_year_X:
            #----Players must have one enough-PA season
            player_list = list(player_set_m1)
        else:
            #----Players must have 2 continous enough-PA seasons
            player_list = list(player_set_m1.intersection(player_set_m2))
    player_dict[str(year)] = tuple(np.sort(player_list))
    print(f'Year {year}: {len(player_list)} players')

#----Find all qualified players
for year in years:
    if year == years[0]:
        player_list_all = set(player_dict[str(year)])
    else:
        player_list_all = player_list_all.union(set(player_dict[str(year)]))
player_list_all = tuple(player_list_all)
print(f'All year: {len(player_list_all)}')
print(f'All Player-Year:{sum([len(L) for L in player_dict.values()])}')
```

## N-Fold preparation and testing set selection
We are going to perform K-fold cross validation to find the 'best' model and hyperparameters. Here we divide all of the players into 12 folds, then take 2 folds to be testing set.</br>
One thing to be noticed is that we did not use <I>sklearn.model_selection.train_test_split</I> to split data sets. Directly applying train_test_split onto X and y will split samples from the same player (in different seasons) into different sets, which should be independent. So we split the folds player-wise.

```python
def n_fold_by_player_generator(X, y, ind, player_season_list, n, random_seed='auto', first_player_idx=[]):
    import random
    if not random_seed == 'auto':
        random.seed(random_seed)
    ID_list = []
    for i in player_season_list:
        ID_list.append(i[0])
    ID_unique_list = list(set(ID_list))
    M = len(ID_unique_list)
    random.shuffle(ID_unique_list)
    
    X_folds = {}; y_folds = {}; i_folds = {}
    fold_index = {}
    for i in range(n):
        fold_index[str(i)] = []
    for i, j in enumerate(player_season_list):
        this_shuffled_ID = ID_unique_list.index(j[0])
        ###---------如果j[0]為某些特定球員，就強制放到測試集裡
        if j[0] in first_player_idx:
            fold_index[str(test_folds[0])].append(i)
        else:
            fold_index[str(int( n / M * this_shuffled_ID))].append(i)

    for i in range(n):
        X_folds[str(i)] = X[fold_index[str(i)],:]
        y_folds[str(i)] = y[fold_index[str(i)],]
        i_folds[str(i)] = ind[fold_index[str(i)],]

    folds = {'X':X_folds, 'y':y_folds, 'i':i_folds}
    return folds

def fold_extractor(folds, fold_numbers):

    if type(fold_numbers) == int:
        fold_numbers = [fold_numbers]

    X_this = folds['X'][str(fold_numbers[0])]
    y_this = folds['y'][str(fold_numbers[0])]
    I_this = folds['i'][str(fold_numbers[0])]
    if len(fold_numbers) >= 2:
        for L in fold_numbers[1::]:
            X_this = np.vstack((X_this, folds['X'][str(L)]))
            y_this = np.hstack((y_this, folds['y'][str(L)]))
            I_this = np.hstack((I_this, folds['i'][str(L)]))
    
    return X_this, y_this, I_this
```

```python
indices = np.arange(y_raw.shape[0])
folds = n_fold_by_player_generator(X_raw, y_raw, indices,
                                   player_season_list, 12,
                                   random_seed=random_seed,
                                   first_player_idx=first_player_idx)
X_test_r, y_test, ind_test = fold_extractor(folds, test_folds)

testing_player_names = list(set([player_season_list[i][1] for i in folds['i']['0'].tolist() + folds['i']['1'].tolist()]))
training_player_names = list(set([player_season_list[i][1] for i in range(len(player_season_list))]).difference(testing_player_names))
all_player_names = list(set(df[df.player_id.isin(player_list_all)]['player_name']))
```

### Define some functions to evaluate the models

```python
def MAE(yt, yp=None):
    if yp is None:
        yp = yt - yt
    return np.mean(np.abs(yt-yp))

def R2(yt, yp=None):
    if yp is None:
        yp = yt - yt
    return 1 - np.var(yt - yp) / np.var(yt)
```

```python
#-----Correlation between y_predict and y_ground_truth
def linear_coef_and_correlation(x, y, model):
    if str(type(model)).split('\'')[1].split('.')[0] == 'keras':
        y_pred = model.predict(x,verbose=0)
    else:
        y_pred = model.predict(x).reshape(-1, 1)
    LR_model = linear_model.LinearRegression()
    LR_model.fit(y_pred, y)
    y_pred = LR_model.predict(y_pred)
    print(f'Linear Model Slope: {LR_model.coef_[0]:.3f}')
    print(f'Linear Model Intercept: {LR_model.intercept_:.3f}')
    print(f'Mae of prediction: {MAE(y, y_pred):.3f}')
    print(f'Mae of naive guess: {MAE(y):.3f}')

    return LR_model

#---plot scatter of y_predict and y_ground_truth
def plot_scatter_with_yp(axs, yt, yp, name = '', r2= 'None', linear_model = 'None'):
    def acc_count(yt, yp):
        count = 0
        for i in range(len(yt)):
            if yt[i] * yp[i] > 0:
                count += 1
        return count / len(yt)
    r2 = R2(yt, yp)
    s = [2,] * yt.shape[0]
    axs.scatter(yp, yt, s)
    xy_lim = np.max((np.max(np.abs(yp)),np.max(np.abs(yt))))*1.1
    axs.plot([-0.2, 0.2], [-0.2, 0.2],':')
    axs.set_xlabel('Predicted wOBA change')
    axs.set_ylabel('Real wOBA change')
    axs.set_title(name + ' Set')
    axs.set_ylim((-xy_lim, xy_lim))
    axs.set_xlim((-xy_lim, xy_lim))
    axs.plot([-xy_lim, xy_lim], [-xy_lim, xy_lim], 'k:')
    
    acc = acc_count(yt, yp)
    axs.text(-xy_lim*0.8, xy_lim*0.8, f'acc: {acc:.3f}')
    axs.text(-xy_lim*0.8, xy_lim*0.7, f'naive mae: {MAE(yt):.3f}')
    axs.text(-xy_lim*0.8, xy_lim*0.6, f'model mae: {MAE(yt, yp):.3f}')
    if not corr == 'None':
        axs.text(-xy_lim*0.8, xy_lim*0.5, f'R-square: {r2:.3f}')
        
def plot_scatter(axs, xt, yt, model, name = '', r2= 'None', linear_model = 'None', return_yp = False):
    if str(type(model)).split('\'')[1].split('.')[0] == 'keras':
        y = model.predict(xt,verbose=0).reshape(-1, 1)
    else:
        y = model.predict(xt)
    try:
        y = linear_model.predict(y)
    except:
        pass
    plot_scatter_with_yp(axs, yt, y)
    
    if return_yp:
        return y

```

```python
def n_fold_ML_training_and_evaluation(model, output_ys=True):
    n_tests = int(n_fold / 2) - 1
    
    mae_list = []
    for io in range(n_tests):
        #-------------------------Get train and valid sets---------------------
        train_folds = [*range(n_fold)]
        valid_folds = [io*2 + 2, io*2 + 3] #Need to be a list

        for i in test_folds + valid_folds:
            train_folds.remove(i)

        X_valid_r, y_valid, I_valid = fold_extractor(folds, valid_folds)
        X_train_r, y_train, I_train = fold_extractor(folds, train_folds)

        scaler = preprocessing.StandardScaler().fit(X_train_r)
        X_train = scaler.transform(X_train_r)
        X_valid = scaler.transform(X_valid_r)
        #----------------------Train Model and Get MAE----------------------
        model.fit(X_train, y_train)
        y_pred_t = model.predict(X_train)
        y_pred_v = model.predict(X_valid)
        
        if io == 0:
            y_trains = y_train
            y_valids = y_valid
            y_preds_t = y_pred_t
            y_preds_v = y_pred_v
        else:
            y_trains = np.hstack((y_trains, y_train))
            y_valids = np.hstack((y_valids, y_valid))
            y_preds_t = np.hstack((y_preds_t, y_pred_t))
            y_preds_v = np.hstack((y_preds_v, y_pred_v))
        
        mae_list.append(MAE(y_valid, y_pred_v))
    
    if output_ys:
        return np.mean(mae_list), y_trains, y_valids, y_preds_t, y_preds_v
    else:
        return np.mean(mae_list)
```

```python
def evaluation(y_train, y_valid, y_pred_t, y_pred_v):
    def one_pair(y, yp, ax, name):
        y = np.array(y).reshape((-1,1))
        yp = np.array(yp).reshape((-1,1))
        LR_model = linear_model.LinearRegression()
        LR_model.fit(yp, y)
        
        r2 = R2(y, yp)

        print('='*20 + name + '='*20)
        print(f'Linear Model Slope: {LR_model.coef_[0][0]:.3f}')
        print(f'Linear Model Intercept: {LR_model.intercept_[0]:.3f}')
        print(f'Mae of prediction: {MAE(y, yp):.3f}')
        print(f'Mae of naive guess: {MAE(y):.3f}')
        print(f'R^2: {r2:.3f}')
        plot_scatter_with_yp(ax, y, yp, name=name, linear_model=LR_model)
    
    fig, axes = plt.subplots(1,2)
    fig.set_size_inches(12,5)
    
    one_pair(y_train, y_pred_t, axes[0], name='Train')
    one_pair(y_valid, y_pred_v, axes[1], name='Valid')
```

```python
def DNN_regression_model(L2_factor, drop_factor, train_shape_1, Leaky_ReLu):
    if Leaky_ReLu:
        activation = LeakyReLU()
    else:
        activation = 'relu'
    def Sub_Layers(N, x, dropout=True):
        x = Dense(N, activation=activation,
                  kernel_regularizer = regularizers.l2(L2_factor))(x)
        x = BatchNormalization()(x)
        if dropout:
            x = Dropout(drop_factor)(x)
        return(x)

    inputs = Input(shape = [train_shape_1,])
    x = Sub_Layers(128, inputs)
    x = Sub_Layers(64, x)
    x = Sub_Layers(64, x)
    x = Sub_Layers(32, x)
    x = Sub_Layers(16, x)
    x = Sub_Layers(8, x, dropout=False)
    x = Dense(8, activation=activation, kernel_regularizer = regularizers.l2(L2_factor))(x)
    
    outputs = Dense(1, activation = 'linear')(x)
    model = Model(inputs, outputs)
    model.compile(loss = 'mse', optimizer = 'rmsprop', metrics = ['mae'])
    return(model)
    
```

##### Optimize hyper-parameters using optuna
From the test run, we can see the model is likely to have overfitting problem. Actually we went back to section 7 to add some kernel_regularizers and dropout layers to ease the problem. In this section we will try to optimize the factors of L2_regularizer and dropout layers using a module named optuna.

```python
#--- Iterative take 2 folds to be validation sets, train the model, 
#    then calculate the averaged mae of training and validation sets.
def n_fold_training_with_early_stop(L2_factor, drop_factor, folds, max_epoch=300, batch_size=256):
    callback = tf.keras.callbacks.EarlyStopping(monitor='val_mae', patience=50, verbose=0)
    
    train_shape_1 = folds['X']['0'].shape[1]
    model = DNN_regression_model(L2_factor, drop_factor, train_shape_1, Leaky_ReLu=Leaky_ReLu)
    
    #-------------------------Get train and valid sets---------------------
    first_io = random.randint(0,4)
    io = first_io
    train_folds = [*range(n_fold)]
    valid_folds = [io*2 + 2, io*2 + 3] #Need to be a list
    for i in test_folds + valid_folds:
        train_folds.remove(i)

    X_valid_r, y_valid, I_valid = fold_extractor(folds, valid_folds)
    X_train_r, y_train, I_train = fold_extractor(folds, train_folds)

    scaler = preprocessing.StandardScaler().fit(X_train_r)
    X_train = scaler.transform(X_train_r)
    X_valid = scaler.transform(X_valid_r)
    #-----------------------------------------------------------------------
    this_h = model.fit(X_train, y_train, batch_size=batch_size,
                       epochs=max_epoch, validation_data=(X_valid, y_valid),
                          verbose=0, callbacks=[callback])
    epoch = len(this_h.history['val_mae'])
    v_mae = np.array(this_h.history['val_mae'])
    t_mae = np.array(this_h.history['mae'])
    
    rest_io = [*range(0, 5)]
    rest_io.remove(first_io)
    for io in rest_io:
        model = DNN_regression_model(L2_factor, drop_factor, train_shape_1, Leaky_ReLu=Leaky_ReLu)
        train_folds = [*range(n_fold)]
        valid_folds = [io*2 + 2, io*2 + 3] #Need to be a list
        for i in test_folds + valid_folds:
            train_folds.remove(i)

        X_valid_r, y_valid, I_valid = fold_extractor(folds, valid_folds)
        X_train_r, y_train, I_train = fold_extractor(folds, train_folds)

        scaler = preprocessing.StandardScaler().fit(X_train_r)
        X_train = scaler.transform(X_train_r)
        X_valid = scaler.transform(X_valid_r)

        this_h = model.fit(X_train, y_train, batch_size=batch_size,
                           epochs=epoch, validation_data=(X_valid, y_valid),
                          verbose=0)
        v_mae += np.array(this_h.history['val_mae'])
        t_mae += np.array(this_h.history['mae'])
        #print(io)
    v_mae /= 5.0
    t_mae /= 5.0
    print(epoch)
    return v_mae, t_mae
```

```python
#--- Iterative take 2 folds to be validation sets, train the model, 
#    then calculate the averaged mae of training and validation sets.
def n_fold_training(L2_factor, drop_factor, folds, epoch=300, batch_size=256, print_fold_num=False):
    train_shape_1 = folds['X']['0'].shape[1]
    v_mae = np.zeros(epoch,)
    t_mae = np.zeros(epoch,)
    
    for io in range(5):
        model = DNN_regression_model(L2_factor, drop_factor, train_shape_1, Leaky_ReLu=Leaky_ReLu)
        train_folds = [*range(n_fold)]
        valid_folds = [io*2 + 2, io*2 + 3] #Need to be a list
        for i in test_folds + valid_folds:
            train_folds.remove(i)

        X_valid_r, y_valid, I_valid = fold_extractor(folds, valid_folds)
        X_train_r, y_train, I_train = fold_extractor(folds, train_folds)

        scaler = preprocessing.StandardScaler().fit(X_train_r)
        X_train = scaler.transform(X_train_r)
        X_valid = scaler.transform(X_valid_r)

        this_h = model.fit(X_train, y_train, batch_size=batch_size,
                           epochs=epoch, validation_data=(X_valid, y_valid),
                          verbose=0)
        v_mae += np.array(this_h.history['val_mae'])
        t_mae += np.array(this_h.history['mae'])
        if print_fold_num:
            print(f'Fold {io+1} is completed!')
    v_mae /= 5.0
    t_mae /= 5.0
    return v_mae, t_mae
```

```python
#--- Take hyperparameters provided by optuna to run the function 'n_fold_training',
#--- then output the minimum v_mae
def dnn_training_for_hyperparameter_tuning(trial, folds=folds):
    train_shape_1 = folds['X']['1'].shape[1]
    cfg = {'Log_L2_factor': trial.suggest_float('Log_L2_factor', -9, 0),
           'drop_factor': trial.suggest_float('drop_factor', 0.1, 0.65)}
    L2_factor = 10**cfg['Log_L2_factor']
    drop_factor = cfg['drop_factor']
    #print([L2_factor, drop_factor])
    
    v_mae, t_mae = n_fold_training_with_early_stop(L2_factor, drop_factor, folds)
    
    min_mae = np.sort(v_mae)[10]
    return min_mae
```

```python
def predict(X, model=model, LRM=LRM.predict, scaler=scaler.transform):
    keras = 'keras' in str(type(model))
    if X.shape == 1:
        X = X.reshape(1, -1)
    if keras:
        y = model.predict(scaler(X),verbose=0)
    else:
        y = model.predict(scaler(X))
        y = y.reshape((-1, 1))
    y = LRM(y)
    return y

def ensemble_predict(X, models, weights, LRMs, scaler):
    ensemble_y = np.zeros((X.shape[0],))
    for model_name in models.keys():
        this_weight = weights[model_name]
        this_y = predict(X, models[model_name], LRMs[model_name].predict, scaler)
        ensemble_y += this_y * this_weight
    ensemble_y /= sum(list(weights.values()))
    return ensemble_y
```
